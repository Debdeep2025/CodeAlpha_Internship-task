```python
import tensorflow_datasets as tfds
import tensorflow as tf

# Load EMNIST Letters
(ds_train, ds_test), ds_info = tfds.load(
    'emnist/letters',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True
)

# Preprocessing function
def normalize_img(image, label):
    image = tf.cast(image, tf.float32) / 255.0
    label -= 1  # Convert labels from 1–26 to 0–25 (A–Z)
    return image, label

# Apply preprocessing
ds_train = ds_train.map(normalize_img).cache().shuffle(10000).batch(128).prefetch(tf.data.AUTOTUNE)
ds_test = ds_test.map(normalize_img).batch(128).prefetch(tf.data.AUTOTUNE)
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(26, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(ds_train, validation_data=ds_test, epochs=10)
```

```python
model.evaluate(ds_test)

for image, label in ds_test.take(1):
    pred = model.predict(image[:1])
    predicted_class = tf.argmax(pred[0]).numpy()
    print(f"Actual: {chr(label[0].numpy() + 65)}, Predicted: {chr(predicted_class + 65)}")
```
